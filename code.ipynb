```python
timer=0
startgame= False
result=False
```

Reading the vedio Face camera

```python
detect=HandDetector(maxHands=1)

cap=cv.VideoCapture(0)

while True:
    ret,image=cap.read()
    
    image=cv.resize(image,(222,205))
    hands,img=detect.findHands(image)
    BG_image=cv.imread('Resources/Back Ground.png')
    image_BG=cv.resize(BG_image,(700,400))
    # cv.imshow('vedio',image_BG)
    # cv.imshow("Game",image)


    # the below logic shows how many fingers are up 
    if startgame:
        if hands:
             hand=hands[0]
             fingers=detect.fingersUp(hand)
             print(fingers)

# We put the camera on our background image
    image_BG[114:319,418:640]=image
    cv.imshow('vedio',image_BG)

    key= cv.waitKey(1)
    if key== ord('s'):
        startgame = True

    if ord('q'):
        break
    

   


cv.destroyAllWindows()
cap.release()
```
