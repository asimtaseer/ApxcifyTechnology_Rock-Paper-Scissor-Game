## Rock-Paper-Scissor Game

Step 1: Importing Libraries

cv2 library is for image and Vedio Processing <br>
cvzone for hand detection<br>
time library to add timer in the game when start<br>
random library is for taking a random iamges (AI moves)

```python
import cv2 as cv
import cvzone
import mediapipe as mp
from cvzone.HandTrackingModule import HandDetector
import time
import random
```

Game control Variables

```python
timer=0
startgame= False
result=False
score_AI=0
score_Player=0
```

```python
detect=HandDetector(maxHands=1)
cap=cv.VideoCapture(0)

while True:
    ret,image=cap.read()
    
    image=cv.resize(image,(216,205))
# detecting the hand on camera
    hands,img=detect.findHands(image)

    BG_image=cv.imread('Resources/Back Ground.png')
    image_BG=cv.resize(BG_image,(700,400))

# main game logic
    if startgame:
        if result is False:
            timer= time.time()- initial_time
        # showing the timer on game 
            cv.putText(image_BG,str(int(timer)),(332,200),cv.FONT_HERSHEY_PLAIN,2,(0,0,255),4)
        # for stoping the timer on 3
            if timer>3:
                result= True
                timer=0

            # below condition defines what the player shows  
                if hands:
                     move=None
                     hand=hands[0]
                     fingers=detect.fingersUp(hand)
                     if fingers==[0,0,0,0,0]:
                         move=1
                     if fingers==[1,1,1,1,1]:
                         move=2
                     if fingers==[0,1,1,0,0]:
                         move=3

                    
                     random_num=random.randint(1,3)
                     image_AI=cv.imread(f'Resources/{random_num}.png')
                     image_BG[112:317,416:638]=image_AI

            # Defined game rules and giving scores  


            #  if player wins increase player score
                     if   (move==1 and random_num==3) or \
                          (move==2 and random_num==1) or \
                          (move==3 and random_num==2):
                      score_Player+=1

            #    if AI wins increase AI score
                     if   (move==3 and random_num==1) or \
                          (move==1 and random_num==2) or \
                          (move==2 and random_num==3):
                      score_AI+=1

                    
# We put the camera on our background image
    image_BG[112:317,54:270]=image

  # as While Loop is continiously running we do this for 
  #  permenently dislpaying AI move
    if result:
        image_BG[112:317,416:638]=image_AI

# displaying scores
    cv.putText(image_BG,str(score_Player),(233,106),cv.FONT_HERSHEY_PLAIN,2,(255,255,255),5)
    cv.putText(image_BG,str(score_AI),(602,106),cv.FONT_HERSHEY_PLAIN,2,(255,255,255),5)

# displaying the complete game
    cv.imshow('vedio',image_BG)


    key= cv.waitKey(1) & 0xff
#  User press S to start game 
    if key== ord('s'):
        startgame = True
        initial_time=time.time()
        result=False
# user press Q to quit the game
    if key==ord('q'):
        break
    

   


cv.destroyAllWindows()
cap.release()
```
