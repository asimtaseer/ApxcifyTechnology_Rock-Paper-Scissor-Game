## Rock-Paper-Scissor Game

Step 1: Importing Libraries

```python
import cv2 as cv
import cvzone
```

Reading the vedio Face camera

```python
cap=cv.VideoCapture(0)

while True:
    ret,image=cap.read()
    image=cv.resize(image,(240,300))
    BG_image=cv.imread('Resources/Back Ground.png')
    image_BG=cv.resize(BG_image,(700,400))
    cv.imshow('vedio',image_BG)
    cv.imshow("Game",image)
    

    if cv.waitKey(1) & 0xff==ord('q'):
        break

cv.destroyAllWindows()
cap.release()
```

Lets Read BG image


```python
BG_image=cv.imread('Resources/Back Ground.png')
image=cv.resize(BG_image,(700,400))
cv.imwrite("Resources/SaveBG.png",image)
# cv.imshow('image',image)

cv.waitKey(0)
cv.destroyAllWindows()
```
