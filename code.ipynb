Reading the vedio Face camera

```python
cap=cv.VideoCapture(0)

while True:
    ret,image=cap.read()
    cv.imshow('vedio',image)
    if cv.waitKey(1) & 0xff==ord('q'):
        break

cv.destroyAllWindows()
cap.release()
```
